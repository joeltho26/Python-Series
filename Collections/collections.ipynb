### Default Dict:

```python
from collections import defaultdict

def default_value():
    return "Not available"

def print_details(*args):
    print(args[0])
    print(args[1])
    print("len:", len(args[1]))
    print("-"*50)

print("-"*100)
print("VARIOUS TYPES OF DEFAULT DICT:")
print("-"*100)

dictionary = defaultdict(int)
dictionary.update({'a':1,'b':2,'c':3})
print_details("Default dictionary:",dictionary)

dictionary_float = defaultdict(float)
dictionary_float.update(dictionary)
print("Index get (without 'g' key present):  -> float",dictionary_float["g"])
print("-"*50)

dictionary_str = defaultdict(str)
dictionary_str.update(dictionary)
print("Index get (without 'g' key present):  -> str",dictionary_str["g"])
print("-"*50)

dictionary_list = defaultdict(list)
dictionary_list.update(dictionary)
print("Index get (without key present):  -> list",dictionary_list["g"])
print("-"*50)

dictionary_set = defaultdict(set)
dictionary_set.update(dictionary)
print("Index get (without key present):  -> set",dictionary_set["g"])
print("-"*50)

dictionary_dict = defaultdict(dict)
dictionary_dict.update(dictionary)
print("Index get (without key present):  -> dict",dictionary_dict["g"])
print("-"*50)

dictionary_tuple = defaultdict(tuple)
dictionary_tuple.update(dictionary)
print("Index get (without key present):  -> tuple",dictionary_tuple["g"])
print("-"*50)

dictionary_lambda = defaultdict(lambda: 10)
dictionary_lambda.update(dictionary)
print("Index get (without key present):  -> lambda",dictionary_lambda["g"])
print("-"*50)

dictionary_function = defaultdict(default_value)
dictionary_function.update(dictionary)
print("Index get (without key present):  -> function",dictionary_function["g"])
print("-"*50)

print("-"*100)
print("DEFAULT DICT METHODS:")
print("-"*100)

dictionary.update({'d':1,'e':2,'f':3})
print("Get (with key present):",dictionary.get("a"))
print("-"*50)

print("Get (without key present):",dictionary.get("g"))
print("-"*50)

print("Index get (without key present):",dictionary["g"])
print("-"*50)

print_details("Keys:",dictionary.keys())

print_details("Values:",dictionary.values())

print_details("Items:",dictionary.items())

print("Popped item:",dictionary.popitem())
print_details("Dictionary after last item popped:",dictionary)

print("Pop of c:",dictionary.pop('c'))
print_details("Dictionary after 'c' item popped:",dictionary)

dictionary1 = dictionary.copy()
print_details("Copied dictionary:",dictionary1)

dictionary1.clear()
print_details("Cleared: ", dictionary1)

dictionary.setdefault("z",26)
print_details("Dictionary data after setdefault:", dictionary)

a = {"a","b","c"}
b  = 0
dictionary_new = defaultdict.fromkeys(a,b)
print_details("Dictionary with fromkeys:",dictionary_new)

dictionary.default_factory = float
print(f"Dictionary with default_factory as float: {dictionary['x']}")
print("-"*50)

dictionary.default_factory = str
print(f"Dictionary with default_factory as str: {dictionary['y']}")
print("-"*50)

dictionary.default_factory = list
print(f"Dictionary with default_factory as list: {dictionary['j']}")
print("-"*50)

dictionary.default_factory = set
print(f"Dictionary with default_factory as set: {dictionary['k']}")
print("-"*50)

dictionary.default_factory = dict
print(f"Dictionary with default_factory as dict: {dictionary['l']}")
print("-"*50)

dictionary.default_factory = tuple
print(f"Dictionary with default_factory as tuple: {dictionary['m']}")
print("-"*50)

dictionary.default_factory = lambda: 10
print(f"Dictionary with default_factory as lambda: {dictionary['n']}")
print("-"*50)

dictionary.default_factory = default_value
print(f"Dictionary with default_factory as function: {dictionary['o']}")
print("-"*50)
```

#### Ordered Dict:

```python
from collections import OrderedDict

dictionary = OrderedDict()

dictionary.update({"name":"John", "gender":"male"})
dictionary["age"] = 24
dictionary["height_in_cms"] = 178

print(dictionary)
print("-"*50)
print("Move to End:")
dictionary.move_to_end("name")
print(dictionary)

# The remaining methods inside the ordered dict are very similar to the actual dictionary methods.
```

#### Named Tuple:

```python
from collections import namedtuple

print("Named Tuple using params as String:")
Point = namedtuple('Point', "x,y")
value = Point(x=10,y=20)
print(value.x, value.y)
print("-"*50) 

print("Named Tuple using params as List:")
Point = namedtuple('Point', ["x","y"])
value = Point(1,2)
print(value.x, value.y)
print("-"*50)

print("Convert to dictionary:")
p = value._asdict()
print(p["x"])
print("-"*50)

print("Make:")
d = Point._make(["apple","apple"])
print(d)
print("-"*50)

print("Count:")
e = d.count("apple")
print(e)
print("-"*50)

print("Index:")
f = d.index("apple")
print(f)
print("-"*50)

print("Fields:")
g = d._fields
print(g)
print("-"*50)

print("Field Default:")
p = namedtuple('Account',"type,balance",defaults=[0])
f = p._field_defaults
print(f)
g = p('premium')
print(g)
print("-"*50)

print("Replace:")
f = g._replace(type='standard')
print(f)
print("-"*50)


```

#### ChainMap:

- grouping iterables

```python
from collections import ChainMap

def print_details(*args):
    print(args[0])
    print(args[1])
    print("len:", len(args[1]))
    print("-"*50)

chain_with_values = ChainMap({1:"a",2:"b",3:"c"},{1:"d",2:"e",3:"f"},{1:"g",2:"h",3:"i"})
print_details("ChainMap with values:", chain_with_values)

chain_with_new_dictA = chain_with_values.new_child()
print_details("ChainMapA with new child:", chain_with_new_dictA)

print_details("ChainMapA with new child maps (current context):", chain_with_new_dictA.maps[0])
print_details("ChainMapA with new child maps (root context):", chain_with_new_dictA.maps[-3:])

chain_with_new_dictA['x'] = 1
chain_with_new_dictA['y'] = 2
chain_with_new_dictA['z'] = 3
print_details("ChainMapA with new child (x,y,z):", chain_with_new_dictA)
print_details("ChainMapA with new child (x,y,z) parents:", chain_with_new_dictA.parents)


chain_without_values = ChainMap()
chain_without_new_dictB = chain_without_values.new_child()
print_details("ChainMapB with new child:", chain_without_new_dictB)

chain_without_new_dictB['a'] = 1
chain_without_new_dictB['b'] = 2
chain_without_new_dictB['c'] = 3
print_details("ChainMapB with new child (a,b,c):", chain_without_new_dictB)
print_details("ChainMapB with new child maps (current context):", chain_without_new_dictB.maps[0])
print_details("ChainMapB with new child maps (root context):", chain_without_new_dictB.maps[-1])
print_details("ChainMapB with new child (a,b,c) parents:", chain_without_new_dictB.parents)
```
