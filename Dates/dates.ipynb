### Dates:

```python
from datetime import datetime, timedelta, time, date, timezone, tzinfo
from dateutil import tz
from dateutil.relativedelta import relativedelta
import pytz

def print_details(*args):
    print(args[0])
    print(args[1])
    print("-"*25)

print_details("Current time:",datetime.now())
print_details("Current time in Fiji with UTC format: ", datetime.now(pytz.timezone('Pacific/Fiji'))) 
print_details("Current time with specific timezone", datetime.now())
print_details("Today:", datetime.today())
print_details("Date:", datetime(2022,10,28))
print_details("Datetime with timestamp:",datetime(2022,10,28,4,28,55))
print_details("Date with replace:",datetime.now().replace(month=5))

# To get day, month, year, hour, minute, second, etc
print_details("Datetime with day:", datetime.now().day)
print_details("Datetime with month:", datetime.now().month)
print_details("Datetime with only year:", datetime.now().year)
print_details("Datetime with only hour:", datetime.now().hour)
print_details("Datetime with only minute:", datetime.now().minute)
print_details("Datetime with only second:", datetime.now().second)
print_details("Datetime with only time:", datetime.now().timetz())
print_details("Datetime with only date:", datetime.now().date())

#strptime
print_details("Date with timestamp with strptime:", datetime.strptime(str(datetime(2022,10,28,4,28,55)),"%Y-%m-%d %H:%M:%S"))

# strftime
print_details("Datetime with only year using 'strftime':", datetime.now().strftime("%Y"))
print_details("Datetime with only month number using 'strftime':", datetime.now().strftime("%m"))
print_details("Datetime with only month string using 'strftime':", datetime.now().strftime("%B"))
print_details("Datetime with only month string shorter using 'strftime':", datetime.now().strftime("%b"))
print_details("Datetime with only date using 'strftime':", datetime.now().strftime("%d"))
print_details("Datetime with only week number using 'strftime':", datetime.now().strftime("%W"))
print_details("Datetime with only day of the week using 'strftime':", datetime.now().strftime("%w"))
print_details("Datetime with only day of the week:", datetime.now().isoweekday)
print_details("Datetime with only day of the week string using 'strftime':", datetime.now().strftime("%A"))
print_details("Datetime with only hour using 'strftime':", datetime.now().strftime("%H"))
print_details("Datetime with only minutes using 'strftime':", datetime.now().strftime("%M"))
print_details("Datetime with only seconds using 'strftime':", datetime.now().strftime("%S"))

# astimezone
print_details("Datetime with astimezone", datetime.now().astimezone())
print_details("Datetime with astimezone -> tz (New Zealand)", datetime.now().astimezone(tz=tz.gettz('NZ-CHAT')))
print_details("Datetime with astimezone -> tz (Germany)", datetime.now().astimezone(tz=tz.gettz('Europe/Berlin')))

# tzname
print_details("Datetime with tzname", datetime.now().tzname())
print_details("Datetime with tzname", datetime.now().astimezone().tzname())
print_details("Datetime with tzname", datetime.now().astimezone(tz=pytz.timezone('Europe/Berlin')).tzname())

#timezone
print_details("Current time with UTC format:",datetime.now(tz=timezone.utc))

#timedelta (year and month not possible)
print_details("Timedelta with 10 days:", datetime.now() - timedelta(days=10))
print_details("Timedelta with 2 months:", datetime.now() - timedelta(hours=15))
print_details("Timedelta with 10 minutes:", datetime.now() - timedelta(minutes=10))
print_details("Timedelta with 20 seconds:", datetime.now() - timedelta(seconds=20))
print_details("Timedelta with 12 week:", datetime.now() - timedelta(weeks=12))

#time & date
print_details("Datetime with combine:", datetime.combine(date.today(),time(10,12,45), tzinfo=pytz.timezone('Asia/Tokyo')))

#relativedelta
print_details("Relativedelta with 10 days: -> days", datetime.now() - relativedelta(days=10))
print_details("Relativedelta with 10 months: -> days", datetime.now() - relativedelta(months=10))
print_details("Relativedelta with 10 years: -> days", datetime.now() - relativedelta(years=10))
print_details("Relativedelta with 10 months: -> days", datetime.now() - relativedelta(weeks=10))
print_details("Relativedelta with 10 leapdays: -> days", datetime.now() - relativedelta(leapdays=10))
print_details("Relativedelta with 10 hours: -> days", datetime.now() - relativedelta(hours=10))
print_details("Relativedelta with 10 minutes: -> days", datetime.now() - relativedelta(minutes=10))
print_details("Relativedelta with 10 seconds: -> days", datetime.now() - relativedelta(seconds=10))

print_details("Relativedelta:", relativedelta(datetime.now(),datetime(2022,9,20,22,40,55)))

```
