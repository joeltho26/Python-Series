### Itertools:

```python
from itertools import accumulate, product, count, cycle, combinations
from itertools import combinations_with_replacement, compress, chain
from itertools import dropwhile, filterfalse, groupby, islice
from itertools import pairwise, permutations, repeat, starmap, takewhile

def print_details(*args):
    print(args[0])
    print(args[1])
    print("-"*25)

data = [1,2,3,4]
data2 = [5,6,7,8]
string_data = "abcdefg"

#accumulate:
accum = accumulate(data, func=lambda a,b: a*b)
print_details("Accumulate", list(accum))

#product:
prod = product(data,data2, repeat=1)
print_details("Accumulate", list(prod))

#count:
print("Count:")
for i in count(5,2):
    print(i, end=" ")
    if i == 21:
        break
print("")
print("-"*25)

#cycle:
print("Cycle:")
for i,j in enumerate(cycle([1,2,3])):
    print(j, end=" ")
    if i == 20:
        break
print("")
print("-"*25)

#combinations
print_details("Combinations:",list(combinations(data,2)))

#combinations_with_replacements
print_details("Combinations with replacement:",list(combinations_with_replacement(data,2)))

#groupby
print("Groupby:")
for i,j in enumerate(groupby(data, key=lambda a: a%2==0)):
    print(j[0], list(j[1]), end=" ")
print("")
print("-"*25)

#repeat
print("Repeat:")
for i,j in enumerate(repeat(data, times=3)):
    print(i, j, sep=",", end="\n")
print("")
print("-"*25)

#permutations
print_details("Permutations:",list(permutations(data,2)))

print_details("Permutations of string:",list(permutations(string_data,2)))

#takewhile => if first element returns false then there will be null output.
new_list = list(takewhile(lambda x: x<3, data))
print_details("takeawhile:", new_list)

#starmap
li =[(2, 3), (3, 1), (4, 6), (5, 3), (6, 5), (7, 2)]
print_details("Starmap:", list(starmap(lambda x, y:x + y, li)))

#dropwhile
new_list = list(dropwhile(lambda x: x<3, data))
print_details("dropwhile:", new_list)

#pairwise
new_list = list(pairwise(data))
print_details("pairwise:", new_list)

#filterfalse
new_list = list(filterfalse(lambda x: x < 2, data))
print_details("filterfalse:", new_list)

#islice
new_list = list(islice(string_data, 1, 5, 2))
print_details("islice:", new_list)

print(string_data)
#compress
new_list = list(compress(string_data, [False,True,True]))
print_details("compress:", new_list)

#chain
chain_data = chain(data)
print_details("chain numerals:", list(chain_data))
next = chain_data.__next__()
print(next)

chain_data_new = chain_data.from_iterable(string_data)
print_details("chain string:", list(chain_data_new))


```
