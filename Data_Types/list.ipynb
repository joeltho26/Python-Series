### List:
- Ordered
- Changeable
- Allow Duplicates

```python
data = [1,5,7,2,5,7,2,9,4,8,1,0]
data_string = list(("apple","orange","grape","banana"))

def print_details(*args):
    print(args[0])
    print(args[1])
    # if len(args) > 2:
    if args[0].lower().find("unpacking") > -1:
        for arg in args[2:]:
            print(arg)
    else:
        pass
    print("-"*25)
    
    

# accessing elements
print_details("Accessing element in list by indexing:",data[1])
print_details("Accessing elements in list by slicing:",data[1:5])
print_details("reversing elements in list by slicing:",data_string[::-1])

# length of elements
print_details("length of list:",len(data_string))

# adding elements
print_details("Adding new element into the list:",data_string.append("mango"))
print_details("Adding new elements into the list:",data_string.extend(("berry","cherry","banana")))
data_string.insert(2,"papaya")
print_details("Inserting element into specific index of a list:",data_string)
data_string.insert(2,["Guava","lemon","grape"])
print_details("Inserting elements as a list into specific index of a list:",data_string)

# count elements
print_details("occurence of elements in a list:",data_string.count("banana"))

# element index
print_details("index of element in a list:",data_string.index("banana",6))

# deleting elements
print_details("Remove of element in a list by value:",data_string.remove("banana"))
print_details("Pop of element in a list by index:",data_string.pop(2))
print_details("Pop of last element in a list:",data_string.pop())

# copy list
data_string_new = data_string.copy()
print_details("copy of element in a list by index:",data_string_new)

# clear list
data_string_new.clear()
print_details("clear of element in a list by index:",data_string_new)

# deleting list
del data_string_new
# print_details("delete list:", data_string_new)

# sorting elements
data_string.sort()
print_details("sort of elements (strings) in a list:",data_string)
data.sort()
print_details("sort of elements (numbers) in a list:",data)

data_string.sort(reverse=True)
print_details("reverse sort of elements (strings) in a list:",data_string)
data.sort(reverse=True)
print_details("reverse sort of element (numbers) in a list:",data)

data_string.sort(key=lambda a: len(a)%2==0)
print_details("sort of elements (strings) in a list by key=lambda:",data_string)

def sorting_fn(number):
    if number%2==1:
        return False
    else:
        return True
    
data.sort(key=sorting_fn)
print_details("sort of element (numbers) in a list by key=function:",data)

# reverse elements
data.reverse()
print_details("reverse of element (numbers) in a list",data)

# list comprehension
print_details("list comprehension:",[i.upper() for i in data_string])
print_details("list comprehension with if:", [i.upper() for i in data_string if len(i)%2==0])
print_details("list comprehension with if..else:", [i.upper() if len(i)%2==0 else None for i in data_string])
print_details("list comprehension with 2 for loops:", [i*j for i in range(0,3) for j in range(0,5)])

#addtion of two list
total_list = data + data_string
print_details("Addition of two list:", total_list)

#list extention with multiplication
list1 = [1,2,3,4,5] * 3
print_details("Extension of list by multiplication:", list1)

#unpacking
a,*b,c = data_string
print_details("unpacking of list:",a,b,c)

a,_,*c = data_string
print_details("unpacking of list with _ (ignoring value):",a,c)
```
