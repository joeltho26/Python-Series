### Set:
- Unordered
- No duplicates
- Changeable (add & removal but not via index)

```python
data = [1,5,7,2,5,7,2,9,4,8,1,0]
data_string = set(("apple","orange","grape","banana","apple"))
data_string1 = set(("apple","orange","grape","banana","apple"))
data_string2 = set(("papaya","orange","guava","banana","cherry"))
data_string3 = set(("apple","orange","grape","banana","apple"))
data_string4 = set(("papaya","orange","guava","banana","cherry"))

def print_details(*args):
    print(args[0])
    print(args[1])
    print("-"*25)
    
# adding elements
data_string.add("lemon")
print_details("adding element into set:",data_string)

data_string.update(("lemon","berry","grape"))
print_details("adding elements into set:", data_string)

# removing elements
data_string.remove("lemon")
print_details("removing element from set:",data_string)

value = data_string.discard("lemon")
print_details("discard element from set:",value)

# difference between sets
value = data_string1.difference(data_string2)
print_details("difference from set A,B:",value)

value = data_string1 - data_string2
print_details("difference from set A - B:",value)

data_string1.difference_update(data_string2)
print_details("difference update from set A,B:",data_string1)

data_string2 -= data_string
print_details("difference update from set A - B:",data_string2)

# intersection between sets
value = data_string3.intersection(data_string4)
print_details("intersection from set A,B:",value)

value = data_string3 & data_string4
print_details("intersection from set A & B:",value)

data_string3.intersection_update(data_string4)
print_details("intersection update from set A,B:",data_string3)

data_string4 &= data_string
print_details("intersection update from set A & B:",data_string4)

# union of sets
data_set = data_string4.union(data_string3)

# subset, superset, isdisjoint between sets
data_string1 = set([1,2,3,4])
data_string2 = set([1,2,3,4,5,6])
print_details("subset A & B:",data_string1.issubset(data_string2))
print_details("superset A & B:",data_string2.issubset(data_string1))
print_details("disjoint A & B:",data_string1.isdisjoint(data_string2))

# list comprehension
print_details("list comprehension:",[i.upper() for i in data_string])
print_details("list comprehension with if:", [i.upper() for i in data_string if len(i)%2==0])
print_details("list comprehension with if..else:", [i.upper() if len(i)%2==0 else None for i in data_string])
```
