```python
data = "   Historical data for such zones is still preserved in the source code, \
but it is not included when compiling the database with standard options...."
    
def print_details(*args):
    print(args[0])
    print(args[1])
    print("-"*25)
    
print_details("Slicing, reversing data",data[::-1])
print_details("length of data",len(data))
print_details("2nd index of data",data[2])
print_details("Capitalize of data",data.capitalize())
print_details("lower of data",data.casefold())
print_details("center of data",data.center(200))
print_details("upper of data",data.upper())
print_details("lower of data",data.lower())
print_details("split of data",data.split(" ", 5))
print_details("strip of data",data.strip(" "))
print_details("find of data",data.find("is", 30))
print_details("replace of data",data.replace(" ", "*", 5))
print_details("partition of data",data.partition(","))
print_details("zfill of data",data.zfill(100))
print_details("title of data",data.title())
print_details("swapcase of data",data.swapcase())
print_details("join of data"," ".partition(","))
print_details("count of data",data.count("s"))
print_details("index of data",data.index(","))
print_details("ljust of data",data.ljust(10,"_"))
print_details("startswith of data",data.startswith(" "))
print_details("endswith of data",data.endswith("."))
print("Format data: \n{0}".format("Surprise!"))
print("-"*25)
print_details("translate data:", data.translate({97:67}))
print_details("maketrans with translate data:", data.translate(str.maketrans("H","B")))
print_details("splitlines data:", data.splitlines(True))
print_details("encode data:", data.encode(encoding="ascii", errors="ignore"))

#string check
print_details("alphabet & number or not",data.replace(" ", "").replace(",","").replace(".","").isalpha())
print_details("alphabet & number or not",data.replace(" ", "").replace(",","").replace(".","").isalnum())
print_details("numeric or not",data.isnumeric())
print_details("title or not",data.strip().istitle())
print_details("printable or not",data.isprintable())
print_details("ascii or not",data.isascii())
print_details("lower or not",data.islower())
print_details("upper or not",data.isupper())
print_details("space or not",data.isspace())
print_details("decimal or not",data.isdecimal())
print_details("digit or not",data.isdigit())
print_details("identifer or not",data.replace(" ", "").replace(",","").replace(".","").isidentifier())

#looping through strings
for i in data.strip(".").split(" "):
    print(i)

    
```
