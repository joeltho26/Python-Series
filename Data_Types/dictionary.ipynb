### Dictionary:
- No duplicate keys
- Changeable
- Not ordered by index instead by keys

```python
person = {"name": "John James", "age":24, "gender":"male"}
location = dict(city="Chennai", state="TamilNadu", direction="South")

def print_details(*args):
    print(args[0])
    print(args[1])
    print("-"*25)
    
# updating dictionary
person.update({"nationality":"India"})
print_details("Update person dictionary:", person)

person.update(location)
print_details("Update complete person dictionary:", person)

person['height'] = 175
print_details("Dictionary:", person)

# updating values in dictionary
person["age"] = 25
print_details("Update person 'age' value in dictionary:", person)

# accessing dictionary values
print_details("Get person 'age' value in dictionary using get method:", person.get("age"))
print_details("Get person 'name' value in dictionary:", person["name"])

# removing elements from dictionary
print_details("Pop person 'gender' value in dictionary:", person.pop("gender"))
print_details("Pop item person removing last key,value in dictionary:", person.popitem())
print_details("Dictionary value after popping out values:", person)

# copy dictionary
person2 = person.copy()
print_details("Copy dictionary:", person2)

# clear dictionary
person2.clear()
print_details("Clear dictionary:", person2)

# delete dictionary
del person2
# print_details("Delete dictionary:", person2)

# keys of dictionary
print_details("Keys of dictionary:", person.keys())

# values of dictionary
print_details("Values of dictionary:", person.values())

# items (key,value) of dictionary
print_details("Key,Value pairs of dictionary:", person.items())

# Create dictionary fromkeys()
keys = ["a","b","c"]
values = [1,2,3]
new_dict = dict.fromkeys(keys,values)
print_details("Create dictionary using fromkeys():", new_dict)

# setdefault() in dictionary
person.setdefault("direction","south")
print_details("Dictionary with setdefault():", person["direction"])
print_details("Dictionary:", person)

# list comprehension
print_details("list comprehension:", {i[0].upper():i[1] for i in person.items()})
print_details("list comprehension with if:", {i[0].upper():i[1] for i in person.items() if isinstance(i[1],str)})
print_details("list comprehension with if..else:", {i[0].upper():i[1] if len(i[0])%2==0 else None for i in person.items()})
```
