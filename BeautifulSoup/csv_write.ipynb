## Using GET request to fetch data: 

```python
import requests
from bs4 import BeautifulSoup
import csv

URL = "http://www.values.com/inspirational-quotes"
headers = {"Content-Type": "application/json; charset=utf-8", 
           "User-Agent": "'Mozilla/5.0 (Macintosh; Intel Mac OS X 14_7) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'"}

r = requests.get(url=URL, headers=headers)
print(r.content)
```

## BeautifulSoup & Prettify:

```python
soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

## Using find method:

```python
quotes=[] 
table = soup.find('div', attrs = {'id':'all_quotes'}) 
```

## Using findAll method:

```python
for row in table.findAll('div', attrs = {'class':'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'}):
    quote = {}
    quote['theme'] = row.h5.text
    quote['url'] = row.a['href']
    quote['img'] = row.img['src']
    quote['lines'] = row.img['alt'].split("#")[0].strip()
    quote['author'] = row.img['alt'].split("#")[1].strip("<>").replace("Author:","")
    quotes.append(quote)
```

## Using CSV writer:

```python
filename = 'Data/inspirational_quotes.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f,['theme','url','img','lines','author'])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)
```
